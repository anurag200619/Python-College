```python
num=int(input("Enter an num"))
count = 0
n = abs(num)

# Handle the special case where num is 0
if n == 0:
    count = 1
else:
    while n > 0:
        n //= 10  # Removes the last digit
        count += 1

print(f"Total digits: {count}")

```

```python
def Ev_factorial(n):
    f = 1
    for i in range(2, n+1, 2):
        f *= i
    return f

def od_factorial(n):
    f = 1
    for i in range(1, n+1, 2):
        f *= i
    return f

num = int(input("Enter a number: "))

if num % 2 == 0:
    print(Ev_factorial(num))
else:
    print(od_factorial(num))

       
```
